```python
import numpy as np
m = 0
epoch = 0
m = 6
eps = 0.15
y_in = [1.2, 1.1, 1, 0.9, 0.95, 1.15]
while True:
    epoch += 1
    y_out = []
    for i in range(0, m):
        if y_in[i] >= 0:
            y_out.append(y_in[i])
        else:
            y_out.append(0)
    print("epoch",epoch,":",y_in)
    if sum(np.array(y_out)>0) == 1:
        print('maximum found in unit : ', np.squeeze(np.where(np.array(y_out)>0)[0])+1)
        break
    for i in range(0, m):
        y_in[i] = y_out[i] - (sum(y_out) - y_out[i]) * eps
    if epoch == 100:
        break
```
